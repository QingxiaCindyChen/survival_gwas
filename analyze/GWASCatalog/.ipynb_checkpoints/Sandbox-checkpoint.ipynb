```python
import pandas as pd
import numpy as np
from urllib import request
import urllib, re, gc, requests, json
from pathos.helpers import mp
from itertools import chain
```

```python
GWASCat = pd.read_table('gwas_catalog_v1.0.2-associations.tsv')
GWASCat = GWASCat[~(GWASCat.MAPPED_TRAIT_URI.astype(str)=='nan')]
```

```python
URLadd = 'https://www.ebi.ac.uk/ols/api/ontologies/efo/terms?iri=http://www.ebi.ac.uk/efo/EFO_0000249'
#URLadd = 'https://www.ebi.ac.uk/ols/api/ontologies/efo/terms?iri={0}'.format(URLs[0])
URLget = requests.get(URLadd)
URLget = requests.get(URLadd).json()
icds = URLget['_embedded']['terms'][0]['annotation']['ICD9_definition_citation']
URLget['_embedded']['terms'][0]['annotation']['SNOMEDCT_definition_citation']
```

```python
DictOut = dict()
URLs = list(set(GWASCat.MAPPED_TRAIT_URI))
URLs = URLs[0:100]

for url in URLs:
    #Multiple URLs may exist.. split into list and find all ICDs within
    Suburl = re.split('\,', url)
    Suburl = [re.sub(' ', '', x) for x in Suburl]
    minilist = []
    for suburl in Suburl:
        urlopen = 'https://www.ebi.ac.uk/ols/api/ontologies/efo/terms?iri={0}'.format(suburl)
        urlget = requests.get(urlopen).json()
        try:
            #Ridiculous structure
            icds = urlget['_embedded']['terms'][0]['annotation']['ICD9_definition_citation']
            icds = [re.sub('ICD9\:', '', x) for x in icds]
            minilist.append(icds)
        except:
            continue
    Add = list(set(chain.from_iterable(minilist)))
    DictOut[url] = list(set(Add))
```

```python
#How many unique traits? and how many have ICD even mentioned? how many have OMIM mentioned?
URLs = list(set(GWASCat.MAPPED_TRAIT_URI))
URLs = URLs[0:500]

URLcount = len(URLs)
icdcount = 0
omimcount = 0
snocount = 0
#take sample of 100 since its a long process
for url in URLs:
    #Multiple URLs may exist.. split into list and find all ICDs within
    Suburl = re.split('\,', url)
    Suburl = [re.sub(' ', '', x) for x in Suburl]
    minilist = []
    for suburl in Suburl:
        urlopen = urllib.request.urlopen(suburl)
        results = str(urlopen.read())
        
        #ICDs = re.findall('ICD9:(\d+?\.\d+)', results)
        ICDs = re.findall('ICD', results)

        if len(ICDs)>0:
            icdcount += 1
        #print('ICD', len(ICDs))
        
        #OMIMs = re.findall('OMIM:(\d+)', results)
        OMIMs = re.findall('OMIM', results)

        if len(OMIMs)>0:
            omimcount += 1
        #print('OMIM', len(OMIMs))
        
        #OMIMs = re.findall('SNOMEDCT:(\d+)', results)
        SNOs = re.findall('SNOMEDCT', results)

        if len(SNOs)>0:
            snocount += 1
        #print('SNOMEDs', len(SNOs))
        
        #print('\n')
```

```python
print('ICD %:', icdcount/500, 'OMIM %:', omimcount/500, 'SNOMEDCT %:', snocount/500)
```

```python
#Make sure any mention of ICD is matched with the exact ICD match
#How many unique traits? and how many have ICD even mentioned? how many have OMIM mentioned?
URLs = list(set(GWASCat.MAPPED_TRAIT_URI))
URLs = URLs[0:100]

URLcount = len(URLs)
icdcount = 0
icd9count = 0
#take sample of 100 since its a long process
for url in URLs:
    #Multiple URLs may exist.. split into list and find all ICDs within
    Suburl = re.split('\,', url)
    Suburl = [re.sub(' ', '', x) for x in Suburl]
    minilist = []
    for suburl in Suburl:
        urlopen = urllib.request.urlopen(suburl)
        results = str(urlopen.read())
        
        ICD9s = re.findall('ICD9:(\d+\.?\d+?)', results)
        ICDs = re.findall('ICD', results)
        if len(ICD9s)>0:
            icd9count += 1
        if len(ICDs)>0:
            icdcount += 1
        if len(ICD9s) != len(ICDs):
            print(url, ICD9s, ICDs)
```
